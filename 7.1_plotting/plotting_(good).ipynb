```python
from matplotlib import pyplot as plt
import numpy as np
```

## pyplot interface

```python
x = np.linspace(-np.pi, np.pi, num=100)
cos_x = np.cos(x)
sin_x = np.sin(x)
plt.plot(x, sin_x, label='$\\sin x$')
plt.plot(x, cos_x, label='$\\cos x$')
plt.legend()
```

## OO interface

```python
fig, ax = plt.subplots(figsize=(8, 4), ncols=2, nrows=2, sharey='row')

my_sin = ax[0, 0]
my_cos = ax[0, 1]

my_sin.plot(x, sin_x, label='$\\sin x$')
my_cos.plot(x, cos_x, label='$\\cos x$')
my_sin.fill_between(x, 0, sin_x, alpha=.1)#, color='green')
my_sin.set_title('sin x')
my_sin.legend()
my_cos.fill_between(x, 0, cos_x, alpha=.1)#, color='blue')
my_cos.set_title('cos x')
my_cos.legend()
#my_sin.fill_between(x, 0, cos_x, alpha=.1, color='red')
#can use a for loop for repetitive tasks

plt.suptitle("Trig functions")
my_cos.set_xlim(-np.pi, np.pi)
my_sin.set_xlim(-np.pi, np.pi)
my_sin.set_xlabel('Angle', rotation=90)
my_cos.set_xlabel('Angle', rotation=0)

for a in ax.flatten():
    a.grid(alpha=.5)

plt.tight_layout()
plt.show()
```
