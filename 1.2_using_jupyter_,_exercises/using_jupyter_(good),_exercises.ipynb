```python
%%timeit
squares = [i**2 for i in range(10)]
#print(f"squares: {squares}")
```

"7 runs, 10,000,000 loops each" means that the code was run for 7 rounds with the code being run 10 million times for each round. Each round took an average of 160 nanoseconds to run with the standard deviation of all the timings being 1.16 nanoseconds.

```python
%%timeit
squares = list()
for i in range(10):
    squares.append(i**2)
#print(f"squares: {squares}")    
```

"7 runs, 1,000,000 loops each" means that the code was run for 7 rounds with the code being run 1 million times for each round. Each round took an average of 205 nanoseconds to run with the standard deviation of all the timings being 1.7 nanoseconds.
